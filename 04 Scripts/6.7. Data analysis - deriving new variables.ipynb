### 1. Import libraries and dataset

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt 
import os
```

### 2. Add calculated variables

```python
# Calculate specified indicators per capita
df['gdp_per_capita'] = df['gdp']/df['total_population']
df['emiss_per_capita'] = df['total_emissions']/df['total_population']
df['agric_prod_per_capita'] = df['agric_gross_prod_value']/df['total_population']
df['agric_export_per_capita'] = df['agric_export']/df['total_population']
df['agric_import_per_capita'] = df['agric_import']/df['total_population']
df['agric_balance_of_trade'] = df['agric_export']-df['agric_import']
```

```python
# Check the shape
df.shape
```

```python
# Check the output
df.head(20)
```

```python
# Add a new column
df.loc[df['agric_balance_of_trade'] >=0, 'trade_balance_group'] = 'Positive'
df.loc[df['agric_balance_of_trade'] <0, 'trade_balance_group'] = 'Negative'
```

```python
df['trade_balance_group'].value_counts()
```

```python
# Check the output
df.head()
```

```python
# Create histogram for 'gdp_per_capita' column
hist = df['gdp_per_capita'].plot.hist(bins = 100)
plt.xlabel('GDP per Capita')
plt.ylabel('Count')
plt.title('Distribution of GDP per Capita')
```

```python
# Create histogram for 'emiss_per_capita' column
hist = df['emiss_per_capita'].plot.hist(bins = 100)
plt.xlabel('Emissions per Capita')
plt.ylabel('Count')
plt.title('Distribution of Emissions per Capita')
```

```python
# Create histogram for 'agric_prod_per_capita' column
hist = df['agric_prod_per_capita'].plot.hist(bins = 100)
plt.xlabel('Agriculture Production per Capita')
plt.ylabel('Count')
plt.title('Distribution of Agriculture Production per Capita')
```

```python
# Create histogram for 'agric_export_per_capita' column
hist = df['agric_export_per_capita'].plot.hist(bins = 100)
plt.xlabel('Agriculture Export per Capita')
plt.ylabel('Count')
plt.title('Distribution of Agriculture Export per Capita')
```

```python
# Create histogram for 'agric_import_per_capita' column
hist = df['agric_import_per_capita'].plot.hist(bins = 100)
plt.xlabel('Agriculture Import per Capita')
plt.ylabel('Count')
plt.title('Distribution of Agriculture Import per Capita')
```
